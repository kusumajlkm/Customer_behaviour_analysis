```python
#null value is replaced with the median
df['Review Rating'] = df.groupby("Category")['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns= {'purchase_amount_(usd)' : 'purchase_amt'})
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels= labels)
df[['age','age_group']].head(20)
```

```python
#create a coloumn purchase_freq_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Quarterly' : 90,
    'Monthly' : 30,
    'Bi-weekly' : 14,
    'Annually' : 365,
    'Every 3 months' : 90}
df['purchase_freq_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
